```python
df_company_names = df_import.iloc[:, 9]
df_company_names.shape
```

```python
import matplotlib.pyplot as plt

name_counts = df_company_names.value_counts()
top_10 = name_counts.nlargest(10)

# Reset the index of the series to make it a dataframe
top_10 = top_10.reset_index()
top_10.columns = ['Company Name', 'Number of Job Postings']

# Plot the data as a horizontal bar plot
top_10.plot(kind='barh', x='Company Name', y='Number of Job Postings', color='blue')

# Add the title, x-label, and y-label
plt.title("Top 10 Companies with the most Job Postings", fontsize=10)
plt.xlabel("Number of Job Postings", fontsize=10)
plt.ylabel("Company Name", fontsize=10)

# Rotate the x-axis labels 90 degrees counterclockwise
plt.xticks(fontsize=6)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```
